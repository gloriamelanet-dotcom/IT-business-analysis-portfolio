```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('customer_data.csv')
data.head()
```

```python
data.info()
data.describe()
```

```python
average_rating = data['Rating'].mean()
print("Average Rating:", average_rating)
```

```python
plt.scatter(data['ResponseTime'], data['Rating'])
plt.title('Response Time vs Rating')
plt.xlabel('Response Time (hours)')
plt.ylabel('Customer Rating')
plt.savefig('images/response_vs_rating.png')
plt.show()
```

```python
plt.scatter(data['Complaints'], data['Rating'])
plt.title('Complaints vs Rating')
plt.xlabel('Number of Complaints')
plt.ylabel('Rating')
plt.savefig('images/response_vs_rating.png')
plt.show()
```

```python
resolved_avg = data.groupby('Resolved')['Rating'].mean()
print(resolved_avg)
```

```python
resolved_avg.plot(kind='bar')
plt.title('Resolution vs Rating')
plt.savefig('images/resolution_impact.png')
plt.show()
```
