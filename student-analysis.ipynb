```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/student_data.csv')
data.head()
```

```python
data.info()
data.describe()
```

```python
subject_averages = data[['Math', 'English', 'Science']].mean()
print(subject_averages)
```

```python
subject_averages.plot(kind='bar')
plt.title('Average Marks Per Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.savefig('../images/average_marks.png')
plt.show()
```

```python
data['Total'] = data[['Math', 'English', 'Science']].sum(axis=1)
data.head()
```

```python
top_students = data.sort_values(by='Total', ascending=False)
top_students.head()
```

```python
plt.hist(data['Math'])
plt.title('Distribution of Math Scores')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.savefig('../images/math_distribution.png')
plt.show()
```
